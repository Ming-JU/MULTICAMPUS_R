```
a=10
b=20

#&&, & : 모두 true이면 true이다. 
# && : 앞의 결과가 거짓이면 뒤문장 수행안함 
# & : 앞의 결과 상관없이 끝까지 수행 
a<10 && (b<-b+1)>=20

a<10 & (b<-b+1)>=20
```

```
b
```

```
#||, | : 하나라도  true이면 true이다. 
# || : 앞의 결과가 참이면 뒤문장 수행안함 
# | : 앞의 결과 상관없이 끝까지 수행 
a=10
b=20
a>=10 || (b<-b+1)>=20
b
a>=10 | (b<-b+1)>=20
b
```
