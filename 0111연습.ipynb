```
#13.  아래의 벡터에 대한 계산 결과를 다음과 같은 이름을 갖는 리스트로 만들어라.
x=c(1,3,4,7,11,18,29)
 
#계산결과는 다음과 같다.
#$곱
#[1]  2  6  8 14 22 36 58
 
#$나눔
#[1]  0.5  1.5  2.0  3.5  5.5  9.0 14.5
 
#$루트
#[1] 1.000000 1.732051 2.000000 2.645751 3.316625 4.242641 5.385165

#곱

a= list(x*2)
a

#나눔
b = list(x/2)
b

#루트
c = list(sqrt(x))
c
```
